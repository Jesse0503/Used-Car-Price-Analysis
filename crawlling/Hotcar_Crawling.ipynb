```python
def get_information_dataframe(information_list):
    Column_Names = ['BRAND','MODEL','YEAR','KILO','LOCATION','Certificate','Warranty','PRICE']
    Info = pd.DataFrame(columns=Column_Names)
    for i in information_list:
        first_dict = []
        second_dict = []
        if len(i) == 16:  
            brand_model_1 = i[0].split(' ',1)
            brand_model_2 = i[8].split(' ',1)
            year_kilo_1 = i[1].split(' / ')
            year_kilo_2 = i[9].split(' / ')
            first_dict.append({'BRAND':brand_model_1[0],'MODEL':brand_model_1[1], 'YEAR':year_kilo_1[0],'KILO':year_kilo_1[1], 
                               'LOCATION':i[3], 'Certificate':1, 'Warranty':1,'PRICE':i[2]})
            second_dict.append({'BRAND':brand_model_2[0],'MODEL':brand_model_2[1], 'YEAR':year_kilo_2[0],'KILO':year_kilo_2[1], 
                               'LOCATION':i[11], 'Certificate':1, 'Warranty':1,'PRICE':i[10]})
            Info = Info.append(first_dict)
            Info = Info.append(second_dict)
        elif len(i) == 13:
            brand_model_1 = i[0].split(' ')
            brand_model_2 = i[7].split(' ')
            year_kilo_1 = i[1].split(' / ')
            year_kilo_2 = i[8].split(' / ')
            first_dict.append({'BRAND':brand_model_1[0],'MODEL':brand_model_1[1], 'YEAR':year_kilo_1[0],'KILO':year_kilo_1[1], 
                               'LOCATION':i[3], 'Certificate':1, 'Warranty':1,'PRICE':i[2]})
            second_dict.append({'BRAND':brand_model_2[0],'MODEL':brand_model_2[1], 'YEAR':year_kilo_2[0],'KILO':year_kilo_2[1], 
                               'LOCATION':i[10], 'Certificate':1, 'Warranty':1,'PRICE':i[9]})
            Info = Info.append(first_dict)
            Info = Info.append(second_dict)
        elif len(i) == 10:
            brand_model_1 = i[0].split(' ')
            brand_model_2 = i[6].split(' ')
            year_kilo_1 = i[1].split(' / ')
            year_kilo_2 = i[7].split(' / ')
            first_dict.append({'BRAND':brand_model_1[0],'MODEL':brand_model_1[1], 'YEAR':year_kilo_1[0],'KILO':year_kilo_1[1], 
                               'LOCATION':i[3], 'Certificate':1, 'Warranty':1,'PRICE':i[2]})
            second_dict.append({'BRAND':brand_model_2[0],'MODEL':brand_model_2[1], 'YEAR':year_kilo_2[0],'KILO':year_kilo_2[1], 
                               'LOCATION':i[9], 'Certificate':1, 'Warranty':1,'PRICE':i[8]})
            Info = Info.append(first_dict)
            Info = Info.append(second_dict)
        else:
            pass
    return Info
```
